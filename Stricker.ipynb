```python
import pandas as pd

ST = pd.read_csv("ST_position.csv")

ST.head(10)
```

```python
ST = ST.drop(['Nationality','Overall','Club','International Reputation','Work Rate','Body Type',
                           'Jersey Number','Joined','Loaned From','Contract Valid Until','GKDiving','GKHandling','GKKicking',
                           'GKPositioning','GKReflexes','Release Clause','Positioning'], axis=1)


ST.head(10)
```

```python

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
ST['Preferred Foot'] =label_encoder.fit_transform(ST['Preferred Foot'])
ST['Position'] =label_encoder.fit_transform(ST['Position'])

def convert_height_to_cm(height):
    try:
        feet, inches = map(int, height.split("'"))
        total_inches = feet * 12 + inches
        height_cm = total_inches * 2.54
        return round(height_cm, 2)
    except Exception as e:
        print(f"Error converting height: {e}")
        return None
    
def convert_weight_to_kg(weight):
    try:
        # Remove 'lbs' from the string and convert to float
        weight_kg = float(weight.replace('lbs', '').strip()) * 0.453592
        return round(weight_kg, 2)
    except Exception as e:
        print(f"Error converting weight: {e}")
        return None

# Applying the conversion functions
ST['Height'] = ST['Height'].apply(convert_height_to_cm)
ST['Weight'] = ST['Weight'].apply(convert_weight_to_kg)

# Function to convert Value and Wage
def convert_value_wage(value):
    if value[-1] == 'M':
        return float(value[1:-1]) * 1e6
    elif value[-1] == 'K':
        return float(value[1:-1]) * 1e3
    return float(value[1:])

# Apply conversion
ST['Value'] = ST['Value'].apply(convert_value_wage).astype(int)
ST['Wage'] = ST['Wage'].apply(convert_value_wage).astype(int)

ST.head(10)
```

```python
ST['Passing'] = ST['Crossing']  +ST['ShortPassing'] +ST['LongPassing']

ST['Ball_Control'] =  ST['Dribbling']  +ST['BallControl'] +ST['Composure'] +ST['HeadingAccuracy'] 

ST['Fitness'] = ST['Acceleration'] +ST['SprintSpeed'] +ST['Agility'] +ST['Reactions'] +ST['Balance'] +ST['Jumping'] +ST['Stamina'] +ST['Strength'] +ST['Aggression'] +ST['Vision']

ST['Defensive_abilities'] =ST['Interceptions'] +ST['Marking']  +ST['StandingTackle']  +ST['SlidingTackle'] 

ST['Shooting'] = ST['Curve'] +ST['FKAccuracy'] +ST['Finishing'] +ST['Penalties'] +ST['LongShots'] 

ST = ST [['Name','ID','Passing','Ball_Control','Value','Fitness','Defensive_abilities','Potential','Weight','Shooting','Age']]
ST.head(10)
```

```python
from sklearn.model_selection import train_test_split

x = ST.drop(['Name','ID', 'Potential'], axis=1)
y = ST['Potential']

# Now split the cleaned data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```
