# 1) Centre Back 

```python
import pandas as pd

CB = pd.read_csv("CB_Position.csv")

CB.head(10)
```

```python
CB = CB.drop(['Nationality','Overall','Club','International Reputation','Work Rate','Body Type',
                           'Jersey Number','Joined','Loaned From','Contract Valid Until','GKDiving','GKHandling','GKKicking',
                           'GKPositioning','GKReflexes','Release Clause','Positioning'], axis=1)


CB.head(10)
```

```python

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
CB['Preferred Foot'] =label_encoder.fit_transform(CB['Preferred Foot'])
CB['Position'] =label_encoder.fit_transform(CB['Position'])

def convert_height_to_cm(height):
    try:
        feet, inches = map(int, height.split("'"))
        total_inches = feet * 12 + inches
        height_cm = total_inches * 2.54
        return round(height_cm, 2)
    except Exception as e:
        print(f"Error converting height: {e}")
        return None
    
def convert_weight_to_kg(weight):
    try:
        # Remove 'lbs' from the string and convert to float
        weight_kg = float(weight.replace('lbs', '').strip()) * 0.453592
        return round(weight_kg, 2)
    except Exception as e:
        print(f"Error converting weight: {e}")
        return None

# Applying the conversion functions
CB['Height'] = CB['Height'].apply(convert_height_to_cm)
CB['Weight'] = CB['Weight'].apply(convert_weight_to_kg)

# Function to convert Value and Wage
def convert_value_wage(value):
    if value[-1] == 'M':
        return float(value[1:-1]) * 1e6
    elif value[-1] == 'K':
        return float(value[1:-1]) * 1e3
    return float(value[1:])

# Apply conversion
CB['Value'] = CB['Value'].apply(convert_value_wage).astype(int)
CB['Wage'] = CB['Wage'].apply(convert_value_wage).astype(int)

CB.head(10)
```

```python
CB['Passing'] = CB['Crossing']  +CB['ShortPassing'] +CB['LongPassing']

CB['Ball_Control'] =  CB['Dribbling']  +CB['BallControl'] +CB['Composure'] +CB['HeadingAccuracy'] 

CB['Fitness'] = CB['Acceleration'] +CB['SprintSpeed'] +CB['Agility'] +CB['Reactions'] +CB['Balance'] +CB['Jumping'] +CB['Stamina'] +CB['Strength'] +CB['Aggression'] +CB['Vision']

CB['Defensive_abilities'] =CB['Interceptions'] +CB['Marking']  +CB['StandingTackle']  +CB['SlidingTackle'] 

CB['Shooting'] = CB['Curve'] +CB['FKAccuracy'] +CB['Finishing'] +CB['Penalties'] +CB['LongShots'] 

CB = CB [['Name','ID','Passing','Ball_Control','Value','Fitness','Defensive_abilities','Potential','Weight','Shooting','Age']]
CB.head(10)
```

xgboost model

```python
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from itertools import product
from sklearn.linear_model import Ridge
import joblib

x = CB.drop(['Name','ID', 'Potential'], axis=1)
y = CB['Potential']

# Now split the cleaned data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
XGboost_model = xgb.XGBRegressor(objective ='reg:squarederror')

# Hyperparameter grid
param_grid = {
    'n_estimators': [200],
    'learning_rate': [0.05],
    'max_depth': [ 7],
    'subsample': [0.6],
    'colsample_bytree': [1.0],
    'reg_alpha': [ 0.1],
    'reg_lambda': [ 1.5]
}

# Grid search with cross-validation
grid_search = GridSearchCV(XGboost_model, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)

# XGboost_model.fit(x_train, y_train)

# Best model from grid search
best_XGboost_model = grid_search.best_estimator_

pred_XGboost_y_test=best_XGboost_model.predict(x_test)
pred_XGboost_y_train=best_XGboost_model.predict(x_train)

print(f'XGBoost Test - {r2_score(y_test,pred_XGboost_y_test)}')
print(f'XGBoost Train - {r2_score(y_train,pred_XGboost_y_train)}')
```

```python
RandomForest_Model = RandomForestRegressor()
RandomForest_Model.fit(x_train , y_train)

pred_RandomForest_y_test=RandomForest_Model.predict(x_test)
pred_RandomForest_y_train=RandomForest_Model.predict(x_train)

print(f'Test - {r2_score(y_test,pred_RandomForest_y_test)}')
print(f'Train - {r2_score(y_train,pred_RandomForest_y_train)}')
```

```python
Ridge_Model = Ridge()
Ridge_Model.fit(x_train, y_train)

pred_Ridge_y_test=Ridge_Model.predict(x_test)
pred_Ridge_y_train=Ridge_Model.predict(x_train)

print(f'Test - {r2_score(y_test,pred_Ridge_y_test)}')
print(f'Train - {r2_score(y_train,pred_Ridge_y_train)}')
```

```python
Combine_test_Centre_Back = ((pred_RandomForest_y_test+pred_XGboost_y_test+pred_Ridge_y_test)/3)
Combine_train_Centre_Back= ((pred_RandomForest_y_train+pred_XGboost_y_train+pred_Ridge_y_train)/3)

print(f'Test - {r2_score(y_test,Combine_test_Centre_Back)}')
print(f'Train - {r2_score(y_train,Combine_train_Centre_Back)}')
```

```python
# Create a DataFrame with player ID, actual Position, and predicted Position
result_df = pd.DataFrame({
    'Name': CB['Name'].iloc[x_test.index],
    'ID': CB['ID'].iloc[x_test.index],
    'Actual_Position': y_test,
    'Predicted_Position': Combine_test_Centre_Back
    
})

# Sort by predicted Position in descending order to get the top players
top_players = result_df.sort_values(by='Predicted_Position', ascending=False).head(10)

# Print the top 10 players
print(top_players)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
RB['Preferred Foot'] =label_encoder.fit_transform(RB['Preferred Foot'])
RB['Position'] =label_encoder.fit_transform(RB['Position'])

def convert_height_to_cm(height):
    try:
        feet, inches = map(int, height.split("'"))
        total_inches = feet * 12 + inches
        height_cm = total_inches * 2.54
        return round(height_cm, 2)
    except Exception as e:
        print(f"Error converting height: {e}")
        return None
    
def convert_weight_to_kg(weight):
    try:
        # Remove 'lbs' from the string and convert to float
        weight_kg = float(weight.replace('lbs', '').strip()) * 0.453592
        return round(weight_kg, 2)
    except Exception as e:
        print(f"Error converting weight: {e}")
        return None

# Applying the conversion functions
RB['Height'] = RB['Height'].apply(convert_height_to_cm)
RB['Weight'] = RB['Weight'].apply(convert_weight_to_kg)

# Function to convert Value and Wage
def convert_value_wage(value):
    if value[-1] == 'M':
        return float(value[1:-1]) * 1e6
    elif value[-1] == 'K':
        return float(value[1:-1]) * 1e3
    return float(value[1:])

# Apply conversion
RB['Value'] = RB['Value'].apply(convert_value_wage).astype(int)
RB['Wage'] = RB['Wage'].apply(convert_value_wage).astype(int)

RB.head(10)
```

```python
RB['Passing'] = RB['Crossing']  +RB['ShortPassing'] +RB['LongPassing']

RB['Ball_Control'] =  RB['Dribbling']  +RB['BallControl'] +RB['Composure'] +RB['HeadingAccuracy'] 

RB['Fitness'] = RB['Acceleration'] +RB['SprintSpeed'] +RB['Agility'] +RB['Reactions'] +RB['Balance'] +RB['Jumping'] +RB['Stamina'] +RB['Strength'] +RB['Aggression'] +RB['Vision']

RB['Defensive_abilities'] =RB['Interceptions'] +RB['Marking']  +RB['StandingTackle']  +RB['SlidingTackle'] 

RB['Shooting'] = RB['Curve'] +RB['FKAccuracy'] +RB['Finishing'] +RB['Penalties'] +RB['LongShots'] 

RB = RB [['Name','ID','Passing','Ball_Control','Value','Fitness','Defensive_abilities','Potential','Weight','Shooting','Age']]
RB.head(10)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, KFold, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from itertools import product
from sklearn.linear_model import Ridge
import joblib

x = RB.drop(['Name','ID', 'Potential'], axis=1)
y = RB['Potential']

# Now split the cleaned data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
XGboost_model = xgb.XGBRegressor(objective ='reg:squarederror')

# Hyperparameter grid
param_grid = {
    'n_estimators': [200],
    'learning_rate': [0.05],
    'max_depth': [ 7],
    'subsample': [0.6],
    'colsample_bytree': [1.0],
    'reg_alpha': [ 0.1],
    'reg_lambda': [ 1.5]
}

# Grid search with cross-validation
grid_search = GridSearchCV(XGboost_model, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)

# XGboost_model.fit(x_train, y_train)

# Best model from grid search
best_XGboost_model = grid_search.best_estimator_

pred_XGboost_y_test=best_XGboost_model.predict(x_test)
pred_XGboost_y_train=best_XGboost_model.predict(x_train)

print(f'XGBoost Test - {r2_score(y_test,pred_XGboost_y_test)}')
print(f'XGBoost Train - {r2_score(y_train,pred_XGboost_y_train)}')
```

```python
RandomForest_Model = RandomForestRegressor()
RandomForest_Model.fit(x_train , y_train)

pred_RandomForest_y_test=RandomForest_Model.predict(x_test)
pred_RandomForest_y_train=RandomForest_Model.predict(x_train)

print(f'Test - {r2_score(y_test,pred_RandomForest_y_test)}')
print(f'Train - {r2_score(y_train,pred_RandomForest_y_train)}')
```

```python
Ridge_Model = Ridge()
Ridge_Model.fit(x_train, y_train)

pred_Ridge_y_test=Ridge_Model.predict(x_test)
pred_Ridge_y_train=Ridge_Model.predict(x_train)

print(f'Test - {r2_score(y_test,pred_Ridge_y_test)}')
print(f'Train - {r2_score(y_train,pred_Ridge_y_train)}')
```

```python
Combine_test_Right_Back = ((pred_RandomForest_y_test+pred_XGboost_y_test+pred_Ridge_y_test)/3)
Combine_train_Right_Back= ((pred_RandomForest_y_train+pred_XGboost_y_train+pred_Ridge_y_train)/3)

print(f'Test - {r2_score(y_test,Combine_test_Right_Back)}')
print(f'Train - {r2_score(y_train,Combine_train_Right_Back)}')
```

```python
# Create a DataFrame with player ID, actual Position, and predicted Position
result_df = pd.DataFrame({
    'Name': RB['Name'].iloc[x_test.index],
    'ID': RB['ID'].iloc[x_test.index],
    'Actual_Position': y_test,
    'Predicted_Position': Combine_test_Right_Back
})

# Sort by predicted Position in descending order to get the top players
top_players = result_df.sort_values(by='Predicted_Position', ascending=False).head(10)

# Print the top 10 players
print(top_players)
```

```python
import pandas as pd
import numpy as np
from PIL import Image, ImageTk

# Global variable to store the image reference
bg_image_ref = None

# Dummy function to load predictions for each position
def load_predictions(position):
    # Example data, replace with actual loading of your predictions
    # You should have different methods to get predictions for each position
    players = ["Player" + str(i) for i in range(1, 21)]
    pred_RandomForest_y_test = np.random.uniform(80, 100, size=20)
    pred_XGboost_y_test = np.random.uniform(80, 100, size=20)
    pred_Ridge_y_test = np.random.uniform(80, 100, size=20)
    
    return players, pred_RandomForest_y_test, pred_XGboost_y_test, pred_Ridge_y_test

# Function for each position
def run_combined_test(position):
    # Load predictions based on position
    players, pred_RandomForest_y_test, pred_XGboost_y_test, pred_Ridge_y_test = load_predictions(position)
    
    # Calculate combined predictions
    Combine_test = (pred_RandomForest_y_test + pred_XGboost_y_test + pred_Ridge_y_test) / 3
    
    # Create a DataFrame for easier sorting
    df = pd.DataFrame({
        'Player_Name': players,
        'Predicted_Rating': Combine_test
    })
    
    # Sort by predicted rating and get the top 10 players
    top_players = df.sort_values(by="Predicted_Rating", ascending=False).head(10)
    
    return list(zip(top_players['Player_Name'], top_players['Predicted_Rating']))

# Define functions for each specific position
def run_combined_test_centre_back():
    return run_combined_test("Centre_Back")

def run_combined_test_left_back():
    return run_combined_test("Left_Back")

def run_combined_test_left_forward():
    return run_combined_test("Left_Forward")

def run_combined_test_left_mid():
    return run_combined_test("Left_Mid")

def run_combined_test_right_mid():
    return run_combined_test("Right_Mid")

def run_combined_test_right_back():
    return run_combined_test("Right_Back")

def run_combined_test_right_forward():
    return run_combined_test("Right_Forward")

def run_combined_test_stricker():
    return run_combined_test("Stricker")


# Map each position to the corresponding function that runs the model
position_model_map = {
    "Centre_Back": run_combined_test_centre_back,
    "Left_Back": run_combined_test_left_back,
    "Left_Forward": run_combined_test_left_forward,
    "Left_Mid": run_combined_test_left_mid,
    "Right_Mid": run_combined_test_right_mid,
    "Right_Back": run_combined_test_right_back,
    "Right_Forward": run_combined_test_right_forward,
    "Stricker": run_combined_test_stricker
}
```

```python
# Function to get the top players for a position by running the appropriate model
def get_top_players(position):
    model_function = position_model_map.get(position)
    
    if model_function:
        return model_function()  # Run the model and get top 10 players
    else:
        return [("Invalid position selected", 0)]

# Function to update the result based on the selected position
def update_result():
    selected_position = position_var.get()
    players = get_top_players(selected_position)
    
    # Clear the current result text
    result_text.delete(1.0, tk.END)
    
    # Display top 10 players for the selected position
    result_text.insert(tk.END, f"Top 10 Players for {selected_position}:\n")
    for player, rating in players:
        result_text.insert(tk.END, f"{player}: {rating}\n")



# Create the main window
root = tk.Tk()
root.title("Football Player Prediction")
root.geometry("600x400")  # Set a fixed window size

# Frame for Header
header_frame = tk.Frame(root, bg="lightblue", pady=10)
header_frame.pack(fill=tk.X)

header_label = tk.Label(header_frame, text="Football Player Prediction System", font=("Helvetica", 16, "bold"), bg="lightblue")
header_label.pack()

# Frame for User Input
input_frame = tk.Frame(root, pady=20)
input_frame.pack()

# Create a dropdown (combobox) for selecting positions
position_var = tk.StringVar()
position_combobox = ttk.Combobox(input_frame, textvariable=position_var, width=30)
position_combobox['values'] = ("Centre_Back", "Left_Back", "Left_Forward", "Left_Mid", "Right_Mid", "Right_Back", "Right_Forward", "Stricker")
position_combobox.set("Select Position")
position_combobox.grid(row=0, column=0, padx=10)

# Create a button to trigger the prediction
predict_button = tk.Button(input_frame, text="Get Top 10 Players", command=lambda: update_result(), bg="lightgreen", font=("Helvetica", 12))
predict_button.grid(row=0, column=1, padx=10)

# Frame for Output
output_frame = tk.Frame(root, pady=20)
output_frame.pack()

# Create a text box to display the result
result_text = tk.Text(output_frame, height=15, width=70, wrap=tk.WORD, font=("Helvetica", 12))
result_text.pack()

# Start the Tkinter event loop
root.mainloop()
```
